```python
# Dependencies and Setup
from pathlib import Path
import pandas as pd

# Loading two files.
school_file = Path("Resources/schools_complete.csv")
student_file = Path("Resources/students_complete.csv")

# Reading school and student files in pandas data frame.
school_df = pd.read_csv(school_file)
student_df = pd.read_csv(student_file)

# Combine two files data into a single dataset. Using student data as the main source, and the merge column is 'school_name' in both files.
combined = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
combined.head()
```

## Local Government Area Summary

```python
# Calculate the Totals (Schools and Students)
schools = combined['school_name'].unique() # select all the schools by its name from the dataframe.
school_counts = len(schools) # get the total number of unique schools.
school_counts

student_counts = combined['student_name'].count() # same methods as above.
student_counts

# Calculate the Total Budget
total_budget = sum(combined['budget'].unique()) # same methods as above.
total_budget
```

```python
# Calculate the Average Scores
math_count = combined['maths_score'].count() #get all the numbers of students taking math from the dataframe.
math_count
average_maths_score = combined['maths_score'].sum() / math_count # use the summation value divided by the total number of students who take math.
average_maths_score

average_reading_score = combined['reading_score'].mean() # use mean() to get the average score.
average_reading_score
```

```python
# Calculate the Percentage Pass Rates
passing_maths_count = combined[(combined["maths_score"] >= 50)].count()["student_name"] # select only math score >= 50 from the math_score column, then find the total number of the student who fall in this cirteria.
passing_maths_count
passing_maths_percentage = (passing_maths_count / float(student_counts)) * 100 #use the total divided by the total students who taking math to get percentage.
passing_maths_percentage

passing_reading_count = combined[combined['reading_score'] >= 50].count()['student_name'] #same method as above but for reading_score.
passing_reading_count
passing_reading_percentage = (passing_reading_count / float(student_counts)) * 100 #same
passing_reading_percentage

filt = combined.loc[(combined['maths_score'] >= 50) & (combined['reading_score'] >= 50)] #I created a filter (boolean mask) to select.
filt
passing_maths_reading_count =filt['student_name'].count() # using the filter to find the total number of students pass both.
passing_maths_reading_count

overall_passing_rate = (passing_maths_reading_count / float(student_counts)) * 100 # same method as above to get passing percentage.
overall_passing_rate
```

```python
# Convert to DataFrame
# Creating a new data frame using all the calculations above and assign its column names
area_summary = pd.DataFrame({'Total_Schools': school_counts, 'Total Students': student_counts, 'Total Budget': total_budget, \
                             'Average Maths Score': average_maths_score, 'Average Reading Score': average_reading_score, \
                             '% Passing Maths': passing_maths_percentage, '% Passing Reading': passing_reading_percentage,\
                             '% Overall Passing': overall_passing_rate}, index=[0])
area_summary

# Formatting
area_summary["Total Students"] = area_summary["Total Students"].map("{:,}".format)
area_summary["Total Budget"] = area_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
area_summary
```

## School Summary

```python
# Use the code provided to select the type per school from school_data
school_types = school_df.set_index(["school_name"])["type"] # setting all index for the dataframe using the school names.
school_types

# Calculate the total student count per school from school_data
per_school_counts = combined['school_name'].value_counts() # This returns the number counts of each school's occurances in the whole dataframe.
per_school_counts

# Calculate the total school budget and per capita spending per school from school_data
g = combined.groupby(['school_name']) # grouping the same school together in one
per_school_budget = g['budget'].mean() # select the grouped dataframe especially on budget column and calculate each school's bugets' mean value.
per_school_budget 
per_school_capita = per_school_budget / per_school_counts # use each school's mean budget divide by each school's occurance counts to get per captia.
per_school_capita

# Calculate the average test scores per school from school_data_complete
per_school_maths = g['maths_score'].mean() #same for using group school name to calculate each school students' average math score. 
per_school_maths
per_school_reading = g['reading_score'].mean() # same for average reading score
per_school_reading
```

```python
# Get the students who passed maths and passed reading by creating separate filtered DataFrames from school_data_complete.
school_passing_maths = combined.loc[combined['maths_score'] >= 50] # use .loc to get the dataframe from selected math score column.
school_passing_maths
school_passing_reading = combined.loc[combined['reading_score'] >= 50] # same for getting the reading score dataframe.
school_passing_reading

# Get the students who passed both reading and maths in a separate DataFrame from school_data_complete.
both_pass = (combined['maths_score'] >= 50) & (combined['reading_score'] >= 50) # create a filter to select only for both score >= 50.
passing_maths_and_reading = combined[both_pass] # apply filter to the entire dataframe to get the result.
passing_maths_and_reading
```

```python
#  Calculate the Percentage Pass Rates
# use three groupby all selecting school name column, then select only the student name as the target column to perfrom value check.
per_school_passing_maths = school_passing_maths.groupby('school_name')['student_name'].count()  # this get value of each school's students pass math.
per_school_passing_reading = school_passing_reading.groupby('school_name')['student_name'].count()# this get value of each school students pass reading
overall_passing_rate = passing_maths_and_reading.groupby('school_name')['student_name'].count() # this gets value of both pass.

# calculate the percentages by using the value gained to divide by the total of each school's counts.
per_school_passing_maths_percentage = (per_school_passing_maths / per_school_counts) * 100
per_school_passing_reading_percentage = (per_school_passing_reading / per_school_counts) * 100
overall_passing_percentage = (overall_passing_rate / per_school_counts) * 100
```

```python
# Convert to DataFrame
per_school_summary = pd.DataFrame({'School Type':school_types, 'Total Students':per_school_counts, 'Total School Budget':per_school_budget, \
                                   'Per Student Budget':per_school_capita, 'Average Maths Score':per_school_maths, \
                                   'Average Reading Score':per_school_reading, '% Passing Maths':per_school_passing_maths_percentage, \
                                   '% Passing Reading':per_school_passing_reading_percentage, \
                                   '% Overall Passing':overall_passing_percentage})

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

## Top Performing Schools (By % Overall Passing)

```python
# Sort and show top five schools
top_schools = per_school_summary.sort_values('% Overall Passing', ascending = False)
top_schools.head(5)
```

```python
# Sort and show bottom five schools
bottom_schools = top_schools.tail(5) # It's aready sorted in descending, so the last five will be the bottom.
bottom_schools
```

## Maths Scores by Year

```python
year_nine=combined.loc[combined["year"] == 9]
year_nine_scores = year_nine.groupby('school_name')['maths_score'].mean() #use groupby to combine all the same school together.
year_nine_scores
```

```python
# Create data series of scores by year levels using conditionals
# use .loc to select all the col+rows meet the requirements.
year_nine = combined.loc[combined["year"] == 9]
year_ten = combined.loc[combined["year"] == 10]
year_eleven = combined.loc[combined["year"] == 11]
year_twelve = combined.loc[combined["year"] == 12]

# Group each by school name, then apply .mean() to all to get the average value to each row.
year_nine_scores_math = year_nine.groupby('school_name')['maths_score'].mean()
year_ten_scores_math = year_ten.groupby('school_name')['maths_score'].mean()
year_eleven_scores_math = year_eleven.groupby('school_name')['maths_score'].mean()
year_twelve_scores_math =year_twelve.groupby('school_name')['maths_score'].mean()

# Combine series into single DataFrame
maths_scores_by_year = pd.DataFrame({'Year 9':year_nine_scores_math, 'Year 10':year_ten_scores_math, \
                                     'Year 11':year_eleven_scores_math, 'Year 12':year_twelve_scores_math})

# Minor data wrangling
maths_scores_by_year.index.name = None

# Display the DataFrame
maths_scores_by_year
```

```python
# Group each by school name, same methods as above.
year_nine_scores_reading = year_nine.groupby('school_name')['reading_score'].mean()
year_ten_scores_reading = year_ten.groupby('school_name')['reading_score'].mean()
year_eleven_scores_reading = year_eleven.groupby('school_name')['reading_score'].mean()
year_twelve_scores_reading = year_twelve.groupby('school_name')['reading_score'].mean()

# Combine series into single DataFrame
reading_scores_by_year = pd.DataFrame({'Year 9':year_nine_scores_reading, 'Year 10':year_ten_scores_reading, \
                                     'Year 11':year_eleven_scores_reading, 'Year 12':year_twelve_scores_reading})

# Minor data wrangling
reading_scores_by_year.index.name = None

# Display the DataFrame
reading_scores_by_year
```
